```python
import pandas as pd
import seaborn as sns
```

```python
base = pd.read_csv('dados.csv', sep = ';')
base
```

```python
# Verificar a distribuiÃ§Ã£o do valor pib
sns.histplot(base.PIB, kde=True, bins=10).set(title='PIB')
```

```python
# Verificar a distribuiÃ§Ã£o do valorempenho
sns.histplot(base.VALOREMPENHO, kde=True, bins=10).set(title='VALOREMPENHO')
```

```python
# 10 menores pibs em grÃ¡fico de barras
pib = base.sort_values('PIB').head(10)
pib = pib.iloc[:, 1: 3]
pib
pib.plot.bar(x='MUNICIPIO', y='PIB', color = 'green')
```

```python
# 10 maiores pibs em grÃ¡fico de barras
pib = base.sort_values('PIB').tail(10)
pib = pib.iloc[:, 1: 3]
pib

pib.plot.bar(x = 'MUNICIPIO', y = 'PIB', color = 'purple')
```

```python
# 10 maiores valorempenho em grÃ¡fico de barras
valorempenho = base.sort_values('VALOREMPENHO').head(10)
valorempenho = valorempenho.iloc[:, [1, 3]]
valorempenho

valorempenho.plot.bar(x = 'MUNICIPIO', y= 'VALOREMPENHO', color = 'red')
```

```python
# 10 menores valorempenho em grÃ¡fico de barras
valorempenho = base.sort_values('VALOREMPENHO').tail(10)
valorempenho = valorempenho.iloc[:, [1, 3]]
valorempenho

valorempenho.plot.bar(x = 'MUNICIPIO', y = 'VALOREMPENHO', color = 'orange')
```
