```python
# Criar modelo de regressão linear para prever qual será o investimento inicial necessário de uma franquia
# a partir da taxa anual cobrada pelo franqueador
```

```python
# Sklearn para criar o modelo de regressão e Yellowbrick para visualização residual
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

from yellowbrick.regressor import ResidualsPlot
```

```python
base = pd.read_csv('slr12.csv', sep = ';')
base.head()
```

```python
# Definição de variáveis X e Y, FrqAnual é a variável independente X e CusInic a variável dependente Y
X = base.iloc[:, 0].values
y = base.iloc[:, 1].values
X
```

```python
# Cálculo de correlação entre X e Y
correlation = np.corrcoef(X, y)
correlation

# Correlação positiva e moderada
```

```python
# Formato de matriz de uma coluna
X = X.reshape(-1, 1)

# Criar o modelo e treinamento 
modelo = LinearRegression()

# Fit para que o treinamento seja executado
modelo.fit(X, y)
```

```python
# Visualização dos dados com pontos
plt.scatter(X, y)

# Linha de previsão
plt.plot(X, modelo.predict(X), color = 'red')
```

```python
# Resultado
taxa_anual = int(input('Digite a taxa anual cobrada pelo franqueador: '))
previsao = modelo.predict([[taxa_anual]])
print(f"O investimento inicial é de: {previsao[0]:.2f}")
```
